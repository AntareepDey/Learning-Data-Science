```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=[1,2,3,4,5,6,7]
Y=[1,2,9,16,25,36,59]
```

```python
plt.plot(X,Y)
```

```python
plt.plot(X,Y,'r*--')
```

Scatter Plot

```python
plt.scatter(X,Y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Scatter plot')
```

```python
y_linear=[1,2,3,4,5,6,7]

plt.subplot(2,1,1)
plt.plot(X,y_linear,'r')

plt.subplot(2,1,2)
plt.plot(X,Y,'g')
```

BAR PLOT

```python
x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  
```

Histogram

```python
a = [22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]
plt.hist(a) 
plt.title("histogram") 
```

Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
```
