```python
import seaborn as sns
import numpy as np
```

**Inbuilt Dataset - TIPS**

```python
df=sns.load_dataset('tips')
num_columns=df.select_dtypes(include=[np.number])

df.head()
```

**Creating Correlation and heatmap**

<h6>Always remove catagorical data to generate correlation and heatmap, because catagorical data cannot be correated.Another way to do so would be converting catagorical data into int or float64 data

```python
num_columns.corr()     
```

```python
sns.heatmap(num_columns.corr())
```

**JOINT PLOT**

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

**PAIR PLOT**

```python
sns.pairplot(df,hue='sex')   #hue is used to differentiate between the data
```

**DIST PLOT**

```python
sns.histplot(df['tip'])
```

**CATAGORICAL PLOTS**

Count Plot

```python
sns.countplot(x='sex',data=df)
```

BAR PLOT

```python
sns.barplot(y='total_bill',x='sex',data=df)
```

BOX PLOT

```python
sns.boxplot(x="day",y="total_bill", data=df,hue="day")
```

VIOLIN PLOT

```python
sns.violinplot(x="total_bill", y="day", data=df,hue="day")
```
