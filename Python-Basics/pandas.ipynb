```python
import pandas as pd
import numpy as np
```

```python
#Creating an array to work with 
arr=np.arange(0,25).reshape(5,5)
```

Creating a pandas Dataframe from array

```python
df=pd.DataFrame(arr,index=["Row1","Row2","Row3","Row4","Row5"],columns=['A','B',"C",'D','E'])
df
```

Get top few entries

```python
df.head()
```

Access columns

```python
df[["A","B"]]
```

Save as CSV

```python
df.to_csv('Test.csv')
```

Accessing Elements

```python
#Two methods to do so 

#1. loc
df.loc["Row1"]

#return type is series
```

```python
#2. iloc
df.iloc[1:,0:2]    #same as numpy slicing 

#return type is dataframe
```

Converting Dataframes to arrays

```python
df.iloc[:,:].values
```

Check number of null Values in different columns

```python
df.isnull().sum()
```
